```python
# Import packages
from bs4 import BeautifulSoup
import urllib.request
import requests
```

```python
url1 = 'https://www.yelp.com/biz/zero-degrees-gainesville-3?osq=Ice+Cream'
ourUrl = urllib.request.urlopen(url1)
soup1 = BeautifulSoup(ourUrl,'html.parser')
```

```python
print(soup1.prettify())
```

```python
review=[]  # Create an empty list to store reviews 
for i in soup1.find_all('p',{'class':'lemon--p__373c0__3Qnnj text__373c0__2Kxyz comment__373c0__3EKjH text-color--normal__373c0__3xep9 text-align--left__373c0__2XGa-'}):  
    per_review=i.find('span')  # extract review
    print(per_review.text)
    review.append(per_review.text)  # append review
```
